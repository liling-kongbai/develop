# 四，思考分析

&emsp;&emsp;通过模型的预测，可以发现模型有输出，但全是空格，如果模型没有出错，说明模型基本上只记住了空格，或者说模型认为对于任何字符的预测，字符的下一个字符都是空格。这种现象非常奇怪，手动执行单步预测也是同样的结果。这可能是严重的过拟合造成的。

&emsp;&emsp;通过打印模型和其他步骤的输入和输出，检查各种变量时，并没有发现模型本身有大错误。但对于嵌入层，双向长短期记忆网络层，由于张量的维度大小没有办法进行检查。但观察模型输出时，发现模型输出张量中的某个值非常大，多次运行打印模型的训练过程，预测过程，手动单步预测过程中的模型输出，发现同样的结果。

&emsp;&emsp;在模型的前向传播函数中，最后并没有加上softmax激活函数，通过加入softmax激活函数进行缩放后，查看模型的输出值，同样发现模型输出张量中的某个值非常大，而其他值非常小。这种现象对于文本成模型来说是非常不正确的。由于模型最后一层的全连接神经网络的输出维度为字典词表的长度，因此较大值就代表了与之相对应的字符的索引。如果依据以上分析，模型确实过拟合严重。

&emsp;&emsp;目前仅发现模型输出张量中的某个值非常大，严重拟合某一字符的现象。但一开始并没有考虑词频的问题，后面才加上了统计字符串中空格和各种单词数量的方法。然后发现模型的词频存在非常大的问题，数量最多的空格达到了约六千个，而单词中数量最多的'the'仅有约三百个，同时单词数量大于一百的，仅有七个单词。很多单词仅使用了几次，甚至很单词仅使用了一次。这种情况确实会严重影响模型的训练和预测。

&emsp;&emsp;现在推测可能是由于空格数量太多导致模型记住了空格这个字符，由于数量过多，任何单词后都接上了空格，从而给字符级的语言模型造成一种现象，就是任何其他字符后，都能接空格，把空格当成了万能的，甚至造成空格后的预测仍然是空格。所以再次推测导致这种现象的结果是严重的过拟合。

&emsp;&emsp;通过不太科学和严谨的测试，减少空格的数量后，模型输出的结果将会转变为't'，'h'，'e'和空格的组合的循环，模型记住了类似'liling kongbai the the the the the the the the the the'的现象。将所有单词'the'全部变为空格，同样减少空格的数量后发现模型又记住了了'a'，'n'，'d'和空格，模型的输出转变为'a'，'n'，'d'和空格的组合的循环。所以说模型必然严重记忆了某些特定的模式。

# 五，总结

&emsp;&emsp;通过分析思考，目前推测模型的问题是由于过拟合导致，也间接说明了模型过于简单单调。由于时间原因，并没有继续进一步的实验。受知识储备限制，暂时也并没有发现有效的处理这种问题的方法。

&emsp;&emsp;同时猜测如果将此模型转变为单词级文本生成模型，采用适当的处理方法，如平滑策略，将各种单词的词频平衡或重新设计模型，效果可能会稍好一些，但是仍然要排除空格的影响，因为它们的数量是巨大的。

&emsp;&emsp;不科学严谨的说！！！作者感觉模型有一点问题，或者说模型在计算方面有问题，又或者是在拟合方面有很大缺陷（肯定的），但是说不出来，没有也无法找到，通过实验观察也没有发现。

&emsp;&emsp;最后，作者认为此模型并不适合处理字符级文本生成任务，很容易造成过拟合现象，而且是严重的过拟合，导致模型仅能记住大量的空格或和某些特定的单词的组合。
